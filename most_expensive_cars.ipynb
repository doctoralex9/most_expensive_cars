```python
import pandas as pd
import matplotlib.pyplot as plt






# Read the CSV file into a pandas DataFrame
df = pd.read_csv('/content/cardekho.csv')  # Update with the correct file path

# Display the first few rows of the DataFrame
print("First few rows of the DataFrame:")
print(df.head())

# Filter the DataFrame to include only data for the year 2017
df_2017 = df[df['year'] == 2017]  # Assuming the column name for the year is 'Year'

# Sort the filtered DataFrame by price in descending order
df_2017_sorted = df_2017.sort_values(by='selling_price', ascending=False)


#select the first row (most expensive car)
most_expensive_car_2017 = df_2017_sorted.iloc[0]

#create a new column in the original DataFrame and fill it the most expensive car
df['Most_Expensive_2017'] = most_expensive_car_2017

# Print the cars with the highest prices for 2017
print("\nCars with the highest prices for 2017:")
print(df_2017_sorted.head())


#Extract relevant columns for visualization (e.g., Price, Mileage, Engine)
price = df_2017_sorted['selling_price']
mileage = df_2017_sorted['mileage(km/ltr/kg)']
engine = df_2017_sorted['engine']

#create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(price, mileage, engine, c='blue', marker='o', alpha=0.5)


#set labels and title
ax.set_xlabel('selling_price')
ax.set_ylabel('mileage')
ax.set_zlabel('engine')
ax.set_title('most expensive cars for 2k17')

plt.show
```
